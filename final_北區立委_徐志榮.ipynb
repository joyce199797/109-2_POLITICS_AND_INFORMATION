## 委問政與政見契合度

```python
corpus_x = xu['jieba_results'].values.tolist() 
```

```python
# TF-IDF
# coding:utf-8  

#算字頻
from sklearn.feature_extraction.text import CountVectorizer  

vectorizer = CountVectorizer()  # 計算單字出現在文件中的次數
X = vectorizer.fit_transform(corpus_x) #  inputmust be a list, return a matrix (#of corpus) x (# of word) # 201 x 4835
word = vectorizer.get_feature_names() # column name which is the word
message_column = word

#計算TFIDF
from sklearn.feature_extraction.text import TfidfTransformer  # 計算tfidf

transformer = TfidfTransformer()  
tfidf = transformer.fit_transform(X) # X : corpus x word matrix, output:201 x 4835
```

```python
# #使用LDA演算法
from sklearn.decomposition import LatentDirichletAllocation

LDA = LatentDirichletAllocation(n_components=6, random_state=None)
LDA.fit(X) # 用countrvectorizer() result 去fit 

topic_message = list()
#觀看結果
for i,topic in enumerate(LDA.components_):
    print(f"TOP 20 WORDS PER TOPIC #{i}")
    topic = [vectorizer.get_feature_names()[index] for index in topic.argsort()[-20:]]
    print(topic)
    for i in topic:
        topic_message.append(i)
```

```python
LDA.fit(tfidf) # 用tfidf result 去fit 

#觀看結果
for i,topic in enumerate(LDA.components_):
    print(f"TOP 20 WORDS PER TOPIC #{i}")
    print([vectorizer.get_feature_names()[index] for index in topic.argsort()[-20:]])
```

```python
for i in topic_message:
#for i in topic_:# 文章內容的topic所有字的總表
    print(i,end = ', ')
```
